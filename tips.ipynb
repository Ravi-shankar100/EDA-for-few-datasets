```
 import seaborn as sns
```

```
df = sns.load_dataset("tips")
```

```
df.head()
```

```
df.dtypes
```

```
df.corr()
```

```
##Coorelation between the variables of the dataset
sns.heatmap(df.corr())
```

```
df['smoker'].value_counts()
```

## Plots for the Numerical Variables
##JointPlot,PairPlot,DistPlot

```
##Jointplot used for Univariate Analysis
sns.jointplot(x ='tip',y='total_bill',data = df,kind = 'hex')
```

```
sns.jointplot(x = 'tip',y= 'total_bill',data = df,kind = 'reg')
```

```
##Pair Plot is used for more than two features in the dataset
sns.pairplot(df)
```

```
df['smoker'].value_counts()
```

```
sns.pairplot(df,hue = 'sex')
```

```
##Dist Plot helps us to get the distribution of the features in the dataset
sns.distplot(df['tip'])
```

```
sns.distplot(df['tip'],kde = False, bins= 10)
```

##Plots for the Categorical Variables
##Count plot,Box Plot,Violin Plot,Bar Plot 

```
sns.countplot('sex', data= df)
```

```
sns.countplot(y = 'sex',data = df)
```

```
##Bar Plot
sns.barplot(x = 'total_bill',y= 'sex',data = df)
```

```
sns.barplot(x='smoker',y ='total_bill',data = df)
```

```
##Box plots

sns.boxplot('sex','total_bill',data = df)
```

```
sns.boxplot('day','total_bill',data = df,palette= 'rainbow')
```

```
sns.boxplot(data = df,orient = 'V')
```

```
sns.boxplot(x= 'total_bill',y= 'day',hue= 'smoker',data = df)
```

```
##Violin Plot

sns.violinplot(x= 'total_bill',y= 'day',data= df,palette= 'rainbow')
```
